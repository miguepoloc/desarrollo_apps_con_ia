## 2. Configuración de la clave de API

```python
import os
from pathlib import Path
from dotenv import load_dotenv
import anthropic

# Cargar variables de entorno desde .env
load_dotenv(Path('..') / '.env')

# Verificar que la clave esté configurada
api_key = os.getenv('ANTHROPIC_API_KEY')
if api_key:
    print(f'✅ Clave de API configurada: {api_key[:10]}...')
else:
    print('❌ ANTHROPIC_API_KEY no encontrada. Configura tu archivo .env')
```

## 3. Ejemplo básico: Mensaje simple

```python
client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))

mensaje = client.messages.create(
    model='claude-3-5-haiku-20241022',
    max_tokens=1024,
    messages=[
        {'role': 'user', 'content': '¿Cuáles son las mejores prácticas en el desarrollo de APIs REST?'}
    ]
)

print(mensaje.content[0].text)
```

## 4. Ejemplo: Con prompt de sistema

```python
mensaje = client.messages.create(
    model='claude-3-5-sonnet-20241022',
    max_tokens=2048,
    system='Eres un experto senior en desarrollo de software con 15 años de experiencia. '
           'Proporciona respuestas detalladas, con ejemplos de código cuando sea apropiado, '
           'y menciona posibles gotchas o problemas comunes.',
    messages=[
        {'role': 'user', 'content': 'Explica cómo implementar autenticación JWT en una API de FastAPI.'}
    ]
)

print(mensaje.content[0].text)
```

## 5. Ejemplo: Conversación multi-turno

```python
conversacion = []

def preguntar(pregunta: str):
    conversacion.append({'role': 'user', 'content': pregunta})
    
    respuesta = client.messages.create(
        model='claude-3-5-haiku-20241022',
        max_tokens=1024,
        system='Eres un tutor de programación que explica conceptos de manera clara y concisa.',
        messages=conversacion
    )
    
    texto = respuesta.content[0].text
    conversacion.append({'role': 'assistant', 'content': texto})
    return texto

# Primera pregunta
print('Claude:', preguntar('¿Qué es Docker y para qué sirve?'))
print()

# Segunda pregunta (mantiene el contexto)
print('Claude:', preguntar('¿Cómo lo usaría para desplegar una aplicación FastAPI?'))
```

## 6. Información de uso

```python
mensaje = client.messages.create(
    model='claude-3-5-haiku-20241022',
    max_tokens=512,
    messages=[{'role': 'user', 'content': 'Hola, ¿qué puedes hacer?'}]
)

print(f'Tokens de entrada: {mensaje.usage.input_tokens}')
print(f'Tokens de salida: {mensaje.usage.output_tokens}')
print(f'Motivo de parada: {mensaje.stop_reason}')
print(f'Respuesta: {mensaje.content[0].text}')
```

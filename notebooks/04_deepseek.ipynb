# DeepSeek - Ejemplos de uso

Este notebook muestra cómo usar la API de DeepSeek para crear aplicaciones con IA.

DeepSeek es compatible con la API de OpenAI, por lo que se usa el mismo cliente apuntando a un endpoint diferente.

**Modelos disponibles:**
- `deepseek-chat` (DeepSeek-V3) - Modelo de chat general
- `deepseek-reasoner` (DeepSeek-R1) - Modelo con cadena de pensamiento

**Documentación:** https://platform.deepseek.com/api-docs  
**Obtén tu clave en:** https://platform.deepseek.com/api_keys

## 2. Configuración de la clave de API

```python
import os
from pathlib import Path
from dotenv import load_dotenv
from openai import OpenAI

# Cargar variables de entorno desde .env
load_dotenv(Path('..') / '.env')

# Configurar cliente de DeepSeek
DEEPSEEK_BASE_URL = 'https://api.deepseek.com'

api_key = os.getenv('DEEPSEEK_API_KEY')
if api_key:
    client = OpenAI(api_key=api_key, base_url=DEEPSEEK_BASE_URL)
    print(f'✅ Clave de API configurada: {api_key[:8]}...')
else:
    print('❌ DEEPSEEK_API_KEY no encontrada. Configura tu archivo .env')
```

## 3. Ejemplo básico: Chat con DeepSeek-V3

```python
respuesta = client.chat.completions.create(
    model='deepseek-chat',
    messages=[
        {'role': 'system', 'content': 'Eres un asistente experto en programación.'},
        {'role': 'user', 'content': '¿Cuál es la diferencia entre DeepSeek-V3 y DeepSeek-R1?'}
    ]
)

print(respuesta.choices[0].message.content)
```

## 4. Ejemplo: Razonamiento con DeepSeek-R1

DeepSeek-R1 es un modelo de razonamiento que muestra su cadena de pensamiento (chain-of-thought) antes de dar la respuesta final.

```python
respuesta = client.chat.completions.create(
    model='deepseek-reasoner',
    messages=[
        {'role': 'user', 'content': 'Si tengo una lista de 1000 elementos y quiero encontrar el número más grande, ¿cuál es el algoritmo más eficiente y por qué?'}
    ]
)

# Mostrar la cadena de razonamiento
if respuesta.choices[0].message.reasoning_content:
    print('=== Cadena de pensamiento ===')
    print(respuesta.choices[0].message.reasoning_content)
    print()

print('=== Respuesta final ===')
print(respuesta.choices[0].message.content)
```

## 5. Ejemplo: Generación de código

```python
respuesta = client.chat.completions.create(
    model='deepseek-chat',
    messages=[
        {
            'role': 'system',
            'content': 'Eres un experto en Python y desarrollo backend. '
                       'Genera código limpio, eficiente y bien documentado.'
        },
        {
            'role': 'user',
            'content': 'Implementa una función de caché con decorador en Python que soporte TTL (time-to-live).'
        }
    ]
)

print(respuesta.choices[0].message.content)
```

## 6. Información de uso

```python
respuesta = client.chat.completions.create(
    model='deepseek-chat',
    messages=[{'role': 'user', 'content': 'Hola, ¿cómo estás?'}]
)

print(f'Tokens de entrada: {respuesta.usage.prompt_tokens}')
print(f'Tokens de salida: {respuesta.usage.completion_tokens}')
print(f'Total de tokens: {respuesta.usage.total_tokens}')
print(f'Respuesta: {respuesta.choices[0].message.content}')
```

## 2. Configuración de la clave de API

```python
import os
from pathlib import Path
from dotenv import load_dotenv

# Cargar variables de entorno desde .env
load_dotenv(Path('..') / '.env')

# Verificar que la clave esté configurada
api_key = os.getenv('OPENAI_API_KEY')
if api_key:
    print(f'✅ Clave de API configurada: {api_key[:8]}...')
else:
    print('❌ OPENAI_API_KEY no encontrada. Configura tu archivo .env')
```

## 3. Ejemplo básico: Chat simple

```python
from openai import OpenAI

client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))

respuesta = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {'role': 'system', 'content': 'Eres un asistente útil y amigable que responde en español.'},
        {'role': 'user', 'content': '¿Qué es la inteligencia artificial?'}
    ]
)

print('Respuesta:', respuesta.choices[0].message.content)
```

## 4. Ejemplo: Conversación con historial

```python
historial = [
    {'role': 'system', 'content': 'Eres un experto en programación Python.'},
    {'role': 'user', 'content': '¿Qué es una lista por comprensión?'},
]

# Primera respuesta
respuesta1 = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=historial
)

texto_respuesta1 = respuesta1.choices[0].message.content
print('Asistente:', texto_respuesta1)

# Continuar la conversación
historial.append({'role': 'assistant', 'content': texto_respuesta1})
historial.append({'role': 'user', 'content': 'Dame un ejemplo práctico'})

respuesta2 = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=historial
)

print('\nAsistente:', respuesta2.choices[0].message.content)
```

## 5. Ejemplo: Generación de código

```python
respuesta = client.chat.completions.create(
    model='gpt-4o',
    messages=[
        {
            'role': 'system',
            'content': 'Eres un experto en desarrollo de APIs REST con FastAPI. '
                       'Responde siempre con código limpio y bien comentado.'
        },
        {
            'role': 'user',
            'content': 'Crea una API REST básica con FastAPI que tenga endpoints CRUD para gestionar una lista de tareas (TODO list).'
        }
    ]
)

print(respuesta.choices[0].message.content)
```

## 6. Información de uso (tokens)

```python
respuesta = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[{'role': 'user', 'content': 'Hola, ¿cómo estás?'}]
)

print(f'Tokens de entrada: {respuesta.usage.prompt_tokens}')
print(f'Tokens de salida: {respuesta.usage.completion_tokens}')
print(f'Total de tokens: {respuesta.usage.total_tokens}')
print(f'Respuesta: {respuesta.choices[0].message.content}')
```
